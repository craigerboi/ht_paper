```python
from scipy.stats import norm 
```

```python
def expected_improvement(prediction, standard_dev, best):
    z = (-abs(.8-prediction)+abs(.8-best))/standard_dev
    return (-abs(.8-prediction)+abs(.8-best))*norm.cdf(z) + standard_dev*norm.pdf(z)

def probability_of_improvement(prediction, standard_dev, best):
    z = (-abs(0.8-prediction)+abs(0.8-best))/standard_dev
    return norm.cdf(z)#(-abs(1.23-prediction)+abs(1.23-best))*norm.cdf(z) + standard_dev*norm.pdf(z)
```

```python
# TO DO: Load in data and run the GP with the acquisition function above.
# Need to make holdout sets of Mn, Fe, Co and combos, how quick to convergence?
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel
from sklearn.preprocessing import StandardScaler
```

```python
oh_vals = pickle.load(open("oh_binding_energies.p", "rb"))
racs = pickle.load(open("racs.p", "rb"))
catalyst_names = pickle.load(open("cat_names_for_ml.p", "rb"))
```

```python

def bayes_opt_test(substring_list, opt_val, ei=True):
    holdout_racs = [racs[idx] for idx, cat_name in enumerate(catalyst_names) if any(substring in cat_name for substring in substring_list) and "penta" not in cat_name]# and "hexa" not in cat_name]
    holdout_oh = [oh_vals[idx] for idx, cat_name in enumerate(catalyst_names) if any(substring in cat_name for substring in substring_list) and "penta" not in cat_name]# and "hexa" not in cat_name]
    holdout_names = [catalyst_names[idx] for idx, cat_name in enumerate(catalyst_names) if any(substring in cat_name for substring in substring_list) and "penta" not in cat_name]# and "hexa" not in cat_name]

    train_racs = [racs[idx] for idx, cat_name in enumerate(catalyst_names) if not any(substring in cat_name for substring in substring_list) and "penta" not in cat_name]# and "hexa" not in cat_name]
    train_oh = [oh_vals[idx] for idx, cat_name in enumerate(catalyst_names) if not any(substring in cat_name for substring in substring_list)  and "penta" not in cat_name]# and "hexa" not in cat_name]
    train_names = [catalyst_names[idx] for idx, cat_name in enumerate(catalyst_names) if not any(substring in cat_name for substring in substring_list) and "penta" not in cat_name]# and "hexa" not in cat_name]
    
    len2return = len(holdout_oh)
    best_prediction = 0
    preds_added = []
    good_or_bad = []
    distance_to_best = []
    best_dist_so_far = 10

    target = np.min(abs(np.asarray(holdout_oh) - opt_val))

    while len(holdout_oh)>0:
        sc = StandardScaler()
        scaled_X_train = sc.fit_transform(train_racs)
        scaled_X_test = sc.transform(holdout_racs)

        kernel = ConstantKernel(1.0, (1e-3, 1e3)) * RBF(1, (1e-3, 1e3))
        gp = GaussianProcessRegressor(kernel=kernel, alpha=.01, n_restarts_optimizer=20)
        gp.fit(scaled_X_train, train_oh)

        preds, stds = gp.predict(scaled_X_test, return_std=True)
        if ei:
            acq_fun_vec = [expected_improvement(x, y, best_prediction) for x, y in zip(preds, stds)]
        else:
            acq_fun_vec = [probability_of_improvement(x, y, best_prediction) for x, y in zip(preds, stds)]
        #print(best_prediction)
        #print(holdout_names[np.argmax(acq_fun_vec)], holdout_oh[np.argmax(acq_fun_vec)])
        #for p, s, acq, n in zip(preds, stds, acq_fun_vec, holdout_names):
        #    print(p, s, acq, n)
        recommend_next = holdout_names[np.argmax(acq_fun_vec)]
        oh_val = holdout_oh[np.argmax(acq_fun_vec)]
        if abs(opt_val-oh_val) < abs(opt_val-best_prediction):
            best_prediction = oh_val

        if abs(abs(opt_val-oh_val)-target)<best_dist_so_far:
            best_dist_so_far = abs(abs(opt_val-oh_val)-target)
        distance_to_best.append(best_dist_so_far)

        if abs(.8-oh_val) < np.mean(abs(np.asarray(holdout_oh) - opt_val)):
            good_or_bad.append(1)
        else:
            good_or_bad.append(0)

        train_racs.append(holdout_racs[np.argmax(acq_fun_vec)])
        train_oh.append(holdout_oh[np.argmax(acq_fun_vec)])
        train_names.append(holdout_names[np.argmax(acq_fun_vec)])
        preds_added.append(abs(holdout_oh[np.argmax(acq_fun_vec)]-opt_val))
        del holdout_racs[np.argmax(acq_fun_vec)]
        del holdout_oh[np.argmax(acq_fun_vec)]
        del holdout_names[np.argmax(acq_fun_vec)]
    return distance_to_best, good_or_bad, len2return
    

```

```python
ret_dist, good_or_bad, len_returned = bayes_opt_test(["Fe", "Mn", "Cr"], 0.8)
plt.plot(ret_dist, label="Cr, Mn & Fe, {}".format(len_returned), alpha=0.95)
print(float(sum(good_or_bad))/float(len(good_or_bad)))

ret_dist, good_or_bad, len_returned = bayes_opt_test(["Fe", "Cr"], 0.8) 
plt.plot(ret_dist, label="Cr & Fe, {}".format(len_returned), alpha=0.95)
print(float(sum(good_or_bad))/float(len(good_or_bad)))

ret_dist, good_or_bad, len_returned = bayes_opt_test(["Fe", "Mn"], 0.8) 
plt.plot(ret_dist, label="Mn & Fe, {}".format(len_returned), alpha=0.95)
print(float(sum(good_or_bad))/float(len(good_or_bad)))

ret_dist, good_or_bad, len_returned = bayes_opt_test(["Mn", "Cr"], 0.8) 
plt.plot(ret_dist, label="Cr & Mn, {}".format(len_returned), alpha=0.95)
print(float(sum(good_or_bad))/float(len(good_or_bad)))

ret_dist, good_or_bad, len_returned = bayes_opt_test(["Fe"], 0.8) 
plt.plot(ret_dist, label="Fe, {}".format(len_returned), alpha=0.95)
print(float(sum(good_or_bad))/float(len(good_or_bad)))

ret_dist, good_or_bad, len_returned = bayes_opt_test(["Cr"], 0.8) 
plt.plot(ret_dist, label="Cr, {}".format(len_returned), alpha=0.95)
print(float(sum(good_or_bad))/float(len(good_or_bad)))

ret_dist, good_or_bad, len_returned = bayes_opt_test(["Mn"], 0.8) 
plt.plot(ret_dist, label="Mn, {}".format(len_returned), alpha=0.95)
print(float(sum(good_or_bad))/float(len(good_or_bad)))

plt.legend()
plt.xlabel("Number of evaluations")
plt.ylabel("Distance to best possible $\Delta G_{HO*}$ (eV)")
plt.savefig("bayes_opt.pdf")
plt.show()
```
