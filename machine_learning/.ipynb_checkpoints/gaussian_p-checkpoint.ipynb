```python

rel_logs = pickle.load(open("/home/michael/PycharmProjects/gaussian_job_manage/complete_oh_logs.p", "r"))

valid_cats = dict(pickle.load(open("../molsimp_catalysts/well_behaved_cats.p", "rb")))
valid_cat_names = [x.split("/")[4] for x in valid_cats]
rel_coms = [x.replace(".log", ".com") for x in rel_logs if "VAC" in x]
name2com = defaultdict()
for filename in rel_coms:
    name2com[filename.split("/")[4]] = filename

oh_data = pickle.load(open("/home/michael/PycharmProjects/gaussian_job_manage/oh_binding_energies.p", "r"))
cat_data = pickle.load(open("/home/michael/PycharmProjects/gaussian_job_manage/catalyst_data.p", "r"))
oh_values = []
cat_names = []
actual_rel_coms = []
for name, com in zip(oh_data, rel_coms):
    if name in bad_catalysts or "penta" in name:
        continue
    actual_rel_coms.append(name2com[name])
    oh_values.append(oh_data[name])
    cat_names.append(name)
print(len(oh_values))
```

```python
metal2count = defaultdict(int)
metal2sum = defaultdict(float)
for oh_val, cat_name in zip(oh_values, cat_names):
    #if "hexa" in cat_name:
    #    print(cat_name)
    #    continue
    metal2sum[cat_name[:2]] += oh_val
    metal2count[cat_name[:2]] += 1
metal2avg = defaultdict(float)
for m in metal2count:
    metal2avg[m] = metal2sum[m]/metal2count[m]

errors = []
squared_errs = []
for oh_val, cat_name in zip(oh_values, cat_names):
    errors.append(abs(metal2avg[cat_name[:2]] - oh_val))
    squared_errs.append((metal2avg[cat_name[:2]] - oh_val)**2)
print(np.mean(errors))
print(np.sqrt(np.mean(squared_errs)))
```

```python
def make_rac(comfile, m_depth, l_depth, is_oct):
    #print(comfile)
    with open(comfile, "r") as f:
        lines = f.readlines()
    
    thing = io.read(comfile, format='gaussian')
    properties = ['electronegativity', 'nuclear_charge', 'size', 'topology', 'effective_nuclear_charge']
    
    #thing = Atoms(elements, positions=xyz)
    thing.write("tmp_.xyz")
    first_line = os.popen("head -n3 tmp_.xyz").readlines()[2]
    #ensure we are using the molsimp generated file
    list_of_xyz = re.findall(r"[-+]?\d*\.\d+|\d+", first_line)
    sum_of_xyz = 0.0
    for val in list_of_xyz:
        sum_of_xyz += float(val)
    
    #if sum_of_xyz!=0.0:
    #    return None

    
    this_mol = mol3D() # mol3D instance
    this_mol.readfromxyz("tmp_.xyz")
    feature_names = []
    mc_corrs = np.zeros(shape=(len(properties), (m_depth+1)))
    metal_idx = np.argmax(thing.get_atomic_numbers())
    mc_delta_metricz =  np.zeros(shape=(len(properties), m_depth))
    for idx, p in enumerate(properties):
        delta_list = list(np.asarray(atom_only_deltametric(this_mol, p, m_depth, 0, oct=is_oct)).flatten())
        #print(delta_list)
        del delta_list[0]
        
        mc_corrs[idx] = np.asarray(atom_only_autocorrelation(this_mol, p, m_depth, 0, oct=is_oct)).flatten()
        name_of_idx = ["MC-mult-{}-{}".format(p, x) for x in range(0, m_depth+1)]
        mc_delta_metricz[idx] = delta_list
        feature_names.extend(name_of_idx)
        name_of_idx_diff = ["MC-diff-{}-{}".format(p, x) for x in range(1, m_depth+1)]
        feature_names.extend(name_of_idx_diff)
        
    if is_oct:
        num_connectors = 6
    else:
        num_connectors = 5
    distances = []
    origin = thing.get_positions()[np.argmax(thing.get_atomic_numbers())]
    for xyz in thing.get_positions():
        distances.append(np.sqrt((xyz[0]-origin[0])**2+(xyz[1]-origin[1])**2+(xyz[2]-origin[2])**2))

    nearest_neighbours = np.argpartition(distances, num_connectors)
    nn = [x for x in nearest_neighbours[:num_connectors+1] if x != 0]
    rest_of_autoz = np.zeros(shape=(len(properties), l_depth+1))
    rest_of_deltas = np.zeros(shape=(len(properties), l_depth))
    for idx, p in enumerate(properties):
        rest_of_autoz[idx] = atom_only_autocorrelation(this_mol, p, l_depth, nn, oct=is_oct)
        rest_of_deltas[idx] = atom_only_deltametric(this_mol, p, l_depth, nn)[1:]
        name_of_idx = ["LC-mult-{}-{}".format(p, x) for x in range(0, l_depth+1)]
        name_of_idx_diff = ["LC-diff-{}-{}".format(p, x) for x in range(1, l_depth+1)]
        feature_names.extend(name_of_idx)
        #feature_names.extend(name_of_idx_diff)

    rac_res = np.concatenate((mc_corrs, mc_delta_metricz, rest_of_autoz, rest_of_deltas), axis=None)
    #rac_res = np.concatenate((mc_corrs, mc_delta_metricz, rest_of_deltas), axis=None)
    rac_res = np.concatenate((mc_corrs, mc_delta_metricz, rest_of_autoz, rest_of_deltas), axis=None)
    #print(feature_names)
    return rac_res, feature_names
```

```python
#racs = np.zeros(shape=(len(actual_rel_coms), 42))
racs = []
oh_for_ml = []
cat_names_for_ml = []
for idx, com in enumerate(actual_rel_coms):
    if "31a" or "3a1" or "32a" or "41a" or "hexa" in com:
        is_oct = False
    else:
        is_oct = True
    if "pbp" in com or "hexa" in com or "penta" in com:
        continue
    print(com)
    if make_rac(com, 3, 3, is_oct=is_oct) is not None:
        racs.append(make_rac(com, 3, 1, is_oct=is_oct)[0])
        oh_for_ml.append(oh_values[idx])
        cat_names_for_ml.append(cat_names[idx])
racs = np.asarray(racs)
index_meaning = make_rac(com, 3, 2, is_oct=is_oct)[1]
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.preprocessing import StandardScaler
from sklearn.gaussian_process.kernels import Matern

stdzz = []
preds = []
test_ohs = []
errs = []

loo = LeaveOneOut()
loo.get_n_splits(racs)

for train_index, test_index in loo.split(racs, oh_for_ml):
    sc = StandardScaler()
    X_train, X_test = racs[train_index], racs[test_index]
    y_train, y_test = np.asarray(oh_for_ml)[train_index], np.asarray(oh_for_ml)[test_index]
    scaled_X_train = sc.fit_transform(X_train)
    scaled_X_test = sc.transform(X_test)
    
    kernel = ConstantKernel(1.0, (1e-3, 1e3)) * RBF(1, (1e-3, 1e3))
    gp = GaussianProcessRegressor(kernel=kernel, alpha=.01, n_restarts_optimizer=20)
    gp.fit(scaled_X_train, y_train)
    
    pred, std = gp.predict(scaled_X_test, return_std=True)
    stdzz.append(std[0])
    preds.append(pred[0])
    errs.append(abs(y_test[0]-pred[0]))
    test_ohs.append(y_test[0])
    print(np.asarray(cat_names_for_ml)[test_index][0])
    print("Uncertainty is: {:.3f}".format(std[0]))
    print("Prediction is: {:.3f}".format(pred[0]))
    print("True value is: {:.3f}".format(y_test[0]))
    print("Err is: {:.3f}".format(y_test[0]-pred[0]))

```

```python
print(sqrt(mean_squared_error(np.asarray(test_ohs), np.asarray(preds))))
plt.plot(stdzz, errs, 'ro')
plt.plot([0,1],[0,1])
```
