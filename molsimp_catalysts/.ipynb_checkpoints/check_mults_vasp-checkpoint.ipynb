```python
def check_deviation(oh_val, ooh_val):
    return ooh_val - (0.94*oh_val + 3.275)
```

```python
finished_outcars = pickle.load(open("/home/michael/PycharmProjects/gaussian_job_manage/finished_vasps.p", "rb"))
finished_oohs_perhaps_with_magnetization = finished_outcars[2::3]
outcars_with_lorbit = os.popen("find /home/michael/lorbit_add -name '*OUTCAR'").readlines()
outcars_with_lorbit = [x[:-1] for x in outcars_with_lorbit if "/OOH/" in x]
#find /var/www/html -path "*/myFolder/*.php" -exec do_some_housekeeping {} \;

def has_magnetization_data(outcar):
    with open(outcar, 'r') as f:
        lines = f.readlines()
    for idx, line in enumerate(lines):
        if "NIONS" in line:
            num_atoms = float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[-1])
        if "magnetization (x)" in line:
            return True, idx+4, num_atoms
    
    return False, 0, 0
```

```python
rel_oohs = []
for out in finished_oohs_perhaps_with_magnetization:
    has_mag, start_idx, end_idx = has_magnetization_data(out)
    if has_mag:
        rel_oohs.append(out)
rel_oohs.extend(outcars_with_lorbit)
```

```python
from ase import io

def read_magnetism(outcar):
    structure = io.read(outcar)
    with open(outcar, 'r') as f:
        lines = f.readlines()
    for idx, line in enumerate(lines):
        if "NIONS" in line:
            num_atoms = int(re.findall(r"[-+]?\d*\.\d+|\d+", line)[-1])
        if "magnetization (x)" in line:
            rel_idx = idx+4
            break
    print(rel_idx)
    print(num_atoms)
    total_mags = []
    for line in lines[rel_idx:rel_idx+num_atoms]:
        print(line)
        total_mags.append(float(re.findall(r"[-+]?\d*\.\d+|\d+", line)[-1]))
    return structure, total_mags
```

```python
import numpy as np
cat_to_ooh_oo_colinearity = defaultdict(float)
metal_spin_d = defaultdict(float)
is_ferro = defaultdict(bool)

for l in rel_oohs:
    print(l)
    structure, spin_d = read_magnetism(l)
    distances = []
    positions = structure.get_positions()
    atomic_nums = structure.get_atomic_numbers()
    origin = positions[np.argmax(atomic_nums)]

    for xyz in positions:
        distances.append(np.sqrt((xyz[0]-origin[0])**2+(xyz[1]-origin[1])**2+(xyz[2]-origin[2])**2))
    
    nearest_neighbours = np.argpartition(distances, 6)
    nn = [x for x in nearest_neighbours[:6+1] if x != np.argmax(atomic_nums)]
    atomic_nums = structure.get_atomic_numbers()
    
    for nn_index in nn:
        if atomic_nums[nn_index]==8:
            o_origin = positions[nn_index]
            o_distances = []
            for xyz in positions:
                o_distances.append(np.sqrt((xyz[0]-o_origin[0])**2+(xyz[1]-o_origin[1])**2+(xyz[2]-o_origin[2])**2))
            o_nearest_neighbours = np.argpartition(o_distances, 3)
            nn_o = [x for x in o_nearest_neighbours[:3+1] if x != nn_index]
            for second_nn_index in nn_o:
                if atomic_nums[second_nn_index]==8:
                    first_o_idx = nn_index
                    second_o_idx = second_nn_index
    print(spin_d[first_o_idx]*spin_d[second_o_idx])
    if spin_d[first_o_idx]>0 and spin_d[second_o_idx]>0 and spin_d[0]<0:
        is_ferro[l.split("/")[4]] = False
    elif spin_d[first_o_idx]<0 and spin_d[second_o_idx]<0 and spin_d[0]>0:
        is_ferro[l.split("/")[4]] = False
    else:
        is_ferro[l.split("/")[4]] = True

    cat_to_ooh_oo_colinearity[l.split("/")[4]] = spin_d[first_o_idx]*spin_d[second_o_idx]
    metal_spin_d[l.split("/")[4]] = spin_d[np.argmax(atomic_nums)]
    #if abs(spin_d[first_o_idx])>0.05 and abs(spin_d[second_o_idx])>0.05:
    #    o_o_ratio[l.split("/")[4]] = spin_d[first_o_idx]/spin_d[second_o_idx]
    #cat_to_ooh_oo_colinearity[l.split("/")[4]] = colinearity_oxygens/spin_d[0]
```

```python
name2deviation = defaultdict(float)
for name in beef_oh_values:
    name2deviation[name] = check_deviation(beef_oh_values[name], beef_ooh_values[name])

colinearity = []
deviation = []
metal_s = []
for val in name2deviation:
    if val in cat_to_ooh_oo_colinearity:
        colinearity.append(cat_to_ooh_oo_colinearity[val])
        deviation.append(name2deviation[val])
        metal_s.append(metal_spin_d[val])
        if cat_to_ooh_oo_colinearity[val]>0.01 and metal_spin_d[val]<1. :
            print(val)
            print(name2deviation[val])
            print(metal_spin_d[val])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(9,8))
#ax = plt.axes(projection='3d')
plt.scatter(colinearity, metal_s, alpha=0.75, c=deviation, cmap='seismic');
#ax.view_init(60, 15)

plt.xlabel("spin density(O*O)", fontsize=16)
plt.ylabel("spin density(metal)", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.colorbar()
plt.clim(-.4,.4) 
```
